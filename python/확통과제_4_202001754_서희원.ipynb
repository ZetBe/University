```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy import stats
#2번
#a.
x = np.array([2, 9, 8, 12, 7, 5, 14, 3, 11, 17])
#b.
y = x**2 - 3
print(y)
#c.
correlation = np.corrcoef(x, y)
covariance = np.cov(x, y)[0, 1]
print(correlation)
print(covariance)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy import stats
#3번
x = np.arange(-10, 21)
y = (math.e**(-1*((x-3)**2)/8))/(2*math.sqrt(2*math.pi))
plt.plot(x, y, 'b', label='quick')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy import stats
#4번
x = np.random.uniform(-3, 5, size=1000)
#a.
aver = np.mean(x)
std = np.std(x)
print(aver, std)
#b.
bins = np.arange(-3, 5, 0.25)  # 도수분포구간
hist, bins = np.histogram(x, bins)
plt.hist(x, bins, rwidth=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy import stats
#5번
x = np.random.poisson(7, 500)
#a.
print(np.mean(x))
print(np.std(x))
#b.
bins = np.arange(min(x), max(x))
plt.hist(x, bins, rwidth=0.5)
plt.show()
```
